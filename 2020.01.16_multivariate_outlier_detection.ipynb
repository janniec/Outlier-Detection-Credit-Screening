    B. MCD: Minimum Covariance Determinant (use the mahalanobis distances as the outlier scores)

    A. LOF: Local Outlier Factor

## Age x Debt

```python
feature1 = pairs[0][0]
feature2 = pairs[0][1]
print(feature1, ' x ', feature2)

# https://github.com/yzhao062/pyod/blob/master/notebooks/Compare%20All%20Models.ipynb
```

```python
df = data.copy()
minmax = MinMaxScaler(feature_range=(0, 1))
df[[feature1, feature2]] = minmax.fit_transform(data[[feature1, feature2]])
df[[feature1, feature2]].head()
```

```python
X1 = df[feature1].values.reshape(-1, 1)
X2 = df[feature2].values.reshape(-1, 1)

X = np.concatenate((X1,X2),axis=1)
print(X.shape)
```

```python
# define the number of inliers and outliers
outliers_fraction = 0.01

#Compare given detectors under given settings
#initialize some data
xx, yy = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))

# Define outlier detection tool
random_state = np.random.RandomState(42)
clf = ClusteringBasedLocalOutlierFactor(contamination=outliers_fraction, check_estimator=False, random_state=random_state)

# Fit the model
clf.fit(X)

# Predict the anomaly_score
scores_pred = clf.decision_function(X) * -1

# Prediction of a datapoint category outlier or inlier
y_pred = clf.predict(X)
n_inliers = len(y_pred) - np.count_nonzero(y_pred)
n_outliers = np.count_nonzero(y_pred == 1)
```

```python
df1 = df
df1['outlier'] = y_pred.tolist()
print(df1.shape)
    
# inliers
inliers_feature1 = np.array(df1[feature1][df1['outlier'] == 0]).reshape(-1,1)
inliers_feature2 = np.array(df1[feature2][df1['outlier'] == 0]).reshape(-1,1)
    
# outliers
outliers_feature1 = df1[feature1][df1['outlier'] == 1].values.reshape(-1,1)
outliers_feature2 = df1[feature2][df1['outlier'] == 1].values.reshape(-1,1)
         
print('OUTLIERS:',n_outliers,'INLIERS:',n_inliers)

# Use threshold value to consider a datapoint inlier or outlier
threshold = percentile(scores_pred, 100 * outliers_fraction)  

# decision function calculates the raw anomaly score for every point
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]) * -1
Z = Z.reshape(xx.shape)
```

```python
# plot the levels lines and the points
plt.figure(figsize=(8, 8))

# fill blue map colormap from minimum anomaly score to threshold value
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), threshold, 7),cmap=plt.cm.Blues_r)

# draw red contour line where anomaly score is equal to thresold
a = plt.contour(xx, yy, Z, levels=[threshold],linewidths=2, colors='red')
plt.contourf(xx, yy, Z, levels=[threshold, Z.max()],colors='orange')
# fill orange contour lines where range of anomaly score is from threshold to maximum anomaly score
b = plt.scatter(inliers_feature1, inliers_feature2, c='white',s=20, edgecolor='k')
    
c = plt.scatter(outliers_feature1, outliers_feature2, c='black',s=20, edgecolor='k')
       
plt.axis('tight')   
plt.legend([a.collections[0], b,c], ['learned decision function', 'inliers','outliers'],
           prop=matplotlib.font_manager.FontProperties(size=20),loc='lower right')      
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.title('Cluster-based Local Outlier Factor (CBLOF)')
plt.show();
```
