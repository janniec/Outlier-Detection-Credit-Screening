## CSV

```python
# the column names are arbitrary. 
colnames = arbitrary_categories
data = pd.read_csv(urllib.request.urlopen(path+'crx.data'), delimiter=',', header=None, names=colnames).replace('?', np.NaN)
data 
```

# Explore raw CSV

```python
# should i drop na rows? Sure. Why not.
print(data.shape)
print(data.dropna().shape)
```

```python
data.isnull().sum()
```

```python
# based on the documentation
for index, col in enumerate(data.columns):
    nunique = data[col].nunique()
    if nunique < 20:
        print(index+1, col, '\t', nunique , '\t', data[col].dropna().unique())
    else:
        print(index+1, col, '\t', nunique , '\tCONTINUOUS\t', data[col].dropna().unique().min(), '\t', data[col].dropna().unique().max())
```

```python
data.info()
```

# Transform data

```python
# drop na
data = data.dropna()
print(data.shape)
```

```python
# change continuous 'object' data into numerics
data['Age'] = data['Age'].astype(float)
data['ZipCode'] = data['ZipCode'].astype(int)
```

```python
# dummy binary columns
data = data.replace('t', int(1)).replace('f', int(0)).replace('+', int(1)).replace('-', int(0)).replace('b', int(1)).replace('a', int(0))
```

```python
# dummy multivariate columns
data = pd.get_dummies(data, columns=['Married', 'BankCustomer', 'EducationLevel', 'Ethnicity', 'Citizen'], drop_first=True)
```

# explore transformed CSV

```python
for col in data.columns:
    nunique = data[col].nunique()
    if nunique < 20:
        print(col, '\t', nunique , '\t', data[col].unique())
    else:
        print(col, '\t', nunique , '\tCONTINUOUS\t', data[col].unique().min(), '\t', data[col].unique().max())
```
